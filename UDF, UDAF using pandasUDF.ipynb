```
from pyspark.sql.functions import pandas_udf, PandasUDFType

df = spark.createDataFrame([(1, 1.0), (1, 2.0), (2, 3.0), (2, 5.0), (2, 10.0)],("id", "v"))
  
```

```
display(df)
```

```
@pandas_udf("id long, v double", PandasUDFType.GROUPED_MAP)
def normalize(pdf):
  v = pdf.v
  return pdf.assign(v=(v - v.mean()) / v.std())

df.groupBy("id").apply(normalize).show()
```

```
df = spark.createDataFrame([(1, 1.0), (1, 2.0), (2, 3.0), (2, 5.0), (2, 10.0)],("id", "v"))

@pandas_udf('double', PandasUDFType.SCALAR)
def norm(v):
  return v + 10

df.withColumn('a',norm(df.v))
```

```
import pandas as pd
@pandas_udf('int', PandasUDFType.SCALAR)
def func(pdfc):
  return pdfc*pdfc
  
```

```
df = df.withColumn('sqr',func(df.id))
```

```
df = df.withColumn('key',df.sqr%2)
```

```
#UDAF
import numpy as np
@pandas_udf("key int, sqr int", PandasUDFType.GROUPED_MAP)
def func(pdf):
  sqr = pdf.sqr
  return pdf.assign(sqr=(np.sum(sqr)))

df.select(['key','sqr']).groupBy("key").apply(func).distinct().show()
```
